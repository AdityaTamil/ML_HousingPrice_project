# **Data Preprocessing**

```python
df.hist(figsize=(20,10))
```

### Optimizing the Data

Here the variables total_room, total_bedrooms, population, households are skewed (specifically : right skewed). Hence to get a Gaussian Bell Curve, we can use the log of these varaibles instead.

```python
df['total_rooms'] = np.log(df['total_rooms'] + 1) # 1 is added so that we get only non-zero values
df['total_bedrooms'] = np.log(df['total_bedrooms'] + 1)
df['population'] = np.log(df['population'] + 1)
df['households'] = np.log(df['households'] + 1)
```

```python
df.hist(figsize=(20,10))
```

```python
df['ocean_proximity'].value_counts()
```

Now to include the ocean_proximity we can just consider its 5 individual categories as new variables with values either 0 or 1.

```python
new_var = pd.get_dummies(df.ocean_proximity, dtype= int)
new_var
```

```python
df.drop("ocean_proximity", axis=1, inplace=True)
df = df.join(new_var)
df
```

# **Data Preparation**

## Data Splitting

```python
from sklearn.model_selection import train_test_split

x = df.drop("median_house_value", axis=1)
y = df["median_house_value"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state= 50)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

# **Model Building**

## **Linear Regression**

### Training the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

### Applying the model to make predictions

```python
y_lr_pred_train = lr.predict(x_train)
y_lr_pred_test = lr.predict(x_test)
```

```python
print(y_lr_pred_train)
```

```python
y_train
```

```python
y_lr_pred_test
```

### Evaluate Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_pred_train)
lr_train_rmse = np.sqrt(lr_train_mse)
print(lr_train_rmse)
```

```python
lr_train_r2 = r2_score(y_train, y_lr_pred_train)
lr_train_r2
```

```python
lr_test_mse = mean_squared_error(y_test, y_lr_pred_test)
lr_test_rmse = np.sqrt(lr_test_mse)
print(lr_test_rmse)

lr_test_r2 = r2_score(y_test, y_lr_pred_test)
lr_test_r2
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_rmse, lr_train_r2, lr_test_rmse, lr_test_r2])
lr_results.index = ['Model : ','Training RMSE : ', 'Training R2 :','Test RMSE : ', 'Test R2 : ']
lr_results
```

```python
lr_score = lr.score(x_test, y_test)
lr_score
```

## **Random Forest**

### Training the model

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

### Applying the model to make predictions

```python
y_rf_pred_train = rf.predict(x_train)
y_rf_pred_test = rf.predict(x_test)
```

### Evaluate Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_pred_train)
rf_train_rmse = np.sqrt(rf_train_mse)

rf_train_r2 = r2_score(y_train, y_rf_pred_train)

rf_test_mse = mean_squared_error(y_test, y_rf_pred_test)
rf_test_rmse = np.sqrt(rf_test_mse)

rf_test_r2 = r2_score(y_test, y_rf_pred_test)

rf_results = pd.DataFrame(['Random Forest', rf_train_rmse, rf_train_r2, rf_test_rmse, rf_test_r2])
rf_results.index = ['Model : ','Training RMSE : ', 'Training R2 :','Test RMSE : ', 'Test R2 : ']
rf_results
```

```python
rf_score = rf.score(x_test, y_test)
rf_score
```

## **Neural Networks (Multi-layer Perceptron)**

### Training the model

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(hidden_layer_sizes=(100,50,25,10), random_state=50, max_iter=1000)
mlp.fit(x_train, y_train)
```

### Applying the model to make predictions

```python
y_mlp_pred_train = mlp.predict(x_train)
y_mlp_pred_test = mlp.predict(x_test)
```

### Evaluate Model performance

```python
from sklearn.metrics import mean_squared_error, r2_score

mlp_train_mse = mean_squared_error(y_train, y_mlp_pred_train)
mlp_train_rmse = np.sqrt(mlp_train_mse)

mlp_train_r2 = r2_score(y_train, y_mlp_pred_train)

mlp_test_mse = mean_squared_error(y_test, y_mlp_pred_test)
mlp_test_rmse = np.sqrt(mlp_test_mse)

mlp_test_r2 = r2_score(y_test, y_mlp_pred_test)

mlp_results = pd.DataFrame(['Neural Network(MLP)', mlp_train_rmse, mlp_train_r2, mlp_test_rmse, mlp_test_r2])
mlp_results.index = ['Model : ','Training RMSE : ', 'Training R2 :','Test RMSE : ', 'Test R2 : ']
mlp_results
```

```python
mlp_score = mlp.score(x_test, y_test)
mlp_score
```

# Final Results

```python
final_results = pd.concat([lr_results, rf_results, mlp_results], axis=1)
final_results
```

```python
score_results = pd.DataFrame([lr_score, rf_score, mlp_score])
score_results.index = ['Linear Regression', 'Random Forest', 'Neural Network(MLP)']
score_results.columns = ['Score']
score_results = score_results.transpose()
score_results
```
